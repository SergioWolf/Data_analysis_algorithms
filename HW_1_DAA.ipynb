##### 1. ÐŸÐ¾Ð´Ð±ÐµÑ€Ð¸Ñ‚Ðµ ÑÐºÐ¾Ñ€Ð¾ÑÑ‚ÑŒ Ð¾Ð±ÑƒÑ‡ÐµÐ½Ð¸Ñ (eta) Ð¸ ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ð¸Ñ‚ÐµÑ€Ð°Ñ†Ð¸Ð¹

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[ 1,  1],
              [ 1,  1],
              [ 1,  2],
              [ 1,  5],
              [ 1,  3],
              [ 1,  0],
              [ 1,  5],
              [ 1, 10],
              [ 1,  1],
              [ 1,  2]])
y = [45, 55, 50, 59, 65, 35, 75, 80, 50, 60]
```

```python
def calc_mse(y, y_pred):
    err = np.mean((y - y_pred)**2)
    return err
```

```python
n = X.shape[1]

eta = 5e-3
n_iter = 700

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

for i in range(n_iter):
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    for k in range(W.shape[0]):
        W[k] -= eta * (1/n * 2 * X[:, k] @ (y_pred - y))
    if i % 10 == 0:
        #eta /= 1.1
        print(f'Iteration #{i}: W_new = {W}, MSE = {round(err, 2)}, eta = {eta}')
```

```python
n = X.shape[1]

eta = 5e-3 

# ÑÑ‡ÐµÑ‚Ñ‡Ð¸Ðº Ð¸Ñ‚ÐµÑ€Ð°Ñ†Ð¸Ð¹
iter_num = 0

# ÐºÑ€Ð¸Ñ‚ÐµÑ€Ð¸Ð¹ ÑÑ…Ð¾Ð´Ð¸Ð¼Ð¾ÑÑ‚Ð¸ (Ñ€Ð°Ð·Ð½Ð¸Ñ†Ð° Ð²ÐµÑÐ¾Ð², Ð¿Ñ€Ð¸ ÐºÐ¾Ñ‚Ð¾Ñ€Ð¾Ð¹ Ð°Ð»Ð³Ð¾Ñ€Ð¸Ñ‚Ð¼ Ð¾ÑÑ‚Ð°Ð½Ð°Ð²Ð»Ð¸Ð²Ð°ÐµÑ‚ÑÑ)
min_weight_dist = 1e-6

# Ð·Ð°Ð´Ð°Ð´Ð¸Ð¼ Ð½Ð°Ñ‡Ð°Ð»ÑŒÐ½ÑƒÑŽ Ñ€Ð°Ð·Ð½Ð¸Ñ†Ñƒ Ð²ÐµÑÐ¾Ð² Ð±Ð¾Ð»ÑŒÑˆÐ¸Ð¼ Ñ‡Ð¸ÑÐ»Ð¾Ð¼
weight_dist = np.inf

W = np.array([1, 0.5])
print(f'Number of objects = {n} \
       \nLearning rate = {eta} \
       \nInitial weights = {W} \n')

while weight_dist > min_weight_dist:
    y_pred = np.dot(X, W)
    err = calc_mse(y, y_pred)
    W_new = W - eta * (1/n * 2 * np.dot(X.T, y_pred - y))
    weight_dist = np.linalg.norm(W_new - W, ord=2)
    if iter_num % 10 == 0:
        print(f'Iteration #{iter_num}: W_new = {W}, MSE = {round(err,2)}')
    iter_num += 1
    W = W_new
```
